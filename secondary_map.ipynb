```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import cartopy.crs as ccrs

fin_states = xr.open_dataset('states.nc', decode_times=False)
fin_static = xr.open_dataset('staticData_quarterdeg.nc')
### file is from https://luh.umd.edu/LUH2/LUH2_v2h/states.nc
```

```python
ax = plt.subplot(projection=ccrs.Robinson())

secd_frac = fin_states['secdf'].isel(time=-1) / fin_static['fstnf']

fig = secd_frac.isel(lat=slice(240,480)).plot(cmap='YlGn_r', transform=ccrs.PlateCarree(), cbar_kwargs={'orientation':'horizontal', 'label':''})

ax.set_title('Year 2015 Secondary Forest Fraction')
ax.coastlines() 

plt.savefig('secondary_forest.png', dpi=300)
```
